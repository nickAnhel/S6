```python
import random

import config
import utils
from model import HopfieldNetwork
```

```python
dataset = utils.get_dataset()
train_subset = utils.get_train_subset(dataset)

model = HopfieldNetwork(100)
model.train(train_subset)
```

```python
total_accuracy = 0.0

for i in range(100):
    image = dataset[i]
    noisy_image = utils.get_noisy_image(image, noise_level=random.choice([0.2, 0.3]))

    accuracy, recovered_image = model.test(image, noisy_image)
    total_accuracy += accuracy

    if i < 3:
        utils.draw_images(
            f"Sample {i + 1} — Accuracy: {accuracy:.2f}%",
            ["Original", "Noisy", "Recovered"],
            image,
            noisy_image,
            recovered_image,
        )
```

```python
avg_accuracy = total_accuracy / 100
print(f"Average accuracy: {avg_accuracy:.2f}%")
```

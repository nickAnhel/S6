```python
import matplotlib.pyplot as plt


payoffs = {
    "B1": [2.3, 2.5],
    "B2": [1.7, 2.7],
    "B3": [2.9, 1.9],
    "B4": [3.3, 1.7],
}

x = [0, 1]
p1, p2 = 0.4, 0.6
mid_x = p2
mid_y = 2.3

fig, ax = plt.subplots(figsize=(6, 6))

for name, (y1, y2) in payoffs.items():
    y_values = [y1, y2]
    color = "khaki" if name in ["B1", "B4"] else "brown"
    ax.plot(x, y_values, color=color, lw=1)
    ax.text(1.05, y2, f"{name}={y2}", va="center", fontsize=8, color="gray")
    ax.text(-0.15, y1, f"{name}={y1}", va="center", fontsize=8, color="gray")

ax.plot([mid_x, mid_x], [0, mid_y], "k--")
ax.text(mid_x + 0.01, mid_y / 2, f"{mid_y}", color="brown", fontsize=12)

ax.axvline(0, color="black", lw=1)
ax.axvline(1, color="black", lw=1)

ax.axhline(0, color="black", lw=1)

ax.text(0.15, -0.2, f"p2={p2}", fontsize=10, color="brown")
ax.text(0.75, -0.2, f"p1={p1}", fontsize=10, color="brown")

ax.text(-0.05, -0.15, "A1", fontsize=10)
ax.text(1.01, -0.15, "A2", fontsize=10)

ax.set_xlim(-0.2, 1.2)
ax.set_ylim(-0.5, 3.5)
ax.axis("off")

plt.tight_layout()
plt.show()
```
